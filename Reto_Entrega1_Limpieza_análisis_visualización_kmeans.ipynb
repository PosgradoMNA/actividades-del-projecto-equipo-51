````python
#Utilizamos el GeoDataFrame

puntos_en_mapa = gpd.GeoDataFrame(df_coord, geometry="geometry")
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
world = world.set_index("iso_a3")
world.name.unique()
fig, gax = plt.subplots(figsize=(10,10))

#Filtramos solo Mxico
world.query("name == 'Mexico'").plot(ax=gax, edgecolor='black',color='white')
gax.set_xlabel('LATITUD')
gax.set_ylabel('LONGITUD')
gax.spines['top'].set_visible(False)
gax.spines['right'].set_visible(False)

#Graficamos las coordenadas
puntos_en_mapa.plot(ax=gax, color='blue', alpha = 0.5)
puntos_en_mapa
````

````python
from sklearn.cluster import KMeans

#Aplicamos Kmeans

numero_de_closters = range(10,30) 
mi_kmeans = [KMeans(n_clusters=i) for i in numero_de_closters]
Y_axis = df_coord[['LATITUD']]
X_axis = df_coord[['LONGITUD']]
calulo_kmeans = [mi_kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(mi_kmeans))]

#Visualizamos los codos
plt.figure(figsize=(10,6))
plt.plot(numero_de_closters, calulo_kmeans)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
````

````python
X = df_coord[['LONGITUD', 'LATITUD']]
#Ajustamos el nmero de clusters a 25 
kmeans = KMeans(n_clusters=25).fit(X)

#Obtenemos los centroides
centroids = kmeans.cluster_centers_ 
labels = kmeans.predict(X) 
centroides = kmeans.cluster_centers_ 

#Pasamos los centroides a un dataframe
centroides_df = pd.DataFrame(centroides) 
#Asignamos nombres a las columnas de los centroides, convertimos las coordenadas a tipo punto
centroides_df.rename(mapper={0:'LONGITUD', 1: 'LATITUD'}, axis=1, inplace=True) 
centroides_df["geometry"] = list(zip(centroides_df['LONGITUD'], centroides_df['LATITUD']))             
centroides_df["geometry"] = centroides_df["geometry"].apply(Point)                     
puntos_centroides = gpd.GeoDataFrame(centroides_df, geometry="geometry")
centroides_df
````

Con el objetivo de preparar la visualizacin que existe entre la ubicacin geogrfica de los registros de agua y la calidad de agua, debemos encontrar una forma de representar la calidad de agua predominante por cada cluster, para ello, utilizamos la moda. Recogemos la moda de los valores existentes en ```SEMAFORO``` para cada uno de los clsters, lo cual nos indicar cul es el color del semforo predominante en cada regin, y con esto tenemos una primera informacin sobre qu regin del pas tiene mejor calidad de agua.<br><br>
No obstante, la moda por s sola puede ser un poco engaosa debido a que, si la moda de la variable ```SEMAFORO``` en dos clusters nos entrega el mismo valor, podramos asumir que tienen el mismo nivel de calidad de agua; pero la moda solo nos entrega el valor ms frecuenta, mas no nos indica qu nivel de predominancia tiene ese valor. Por ejemplo, si dos clusters tienen 100 registros de agua, y ambos clusters son representados con el color verde, podramos asumir que ambas regiones tienen agua sumamente limpia; sin embargo, pudiera ser el caso que de 100 registros, 50 tengan agua verde, 10 tengan agua amarilla, y 40 tengan agua roja; entonces, realmente no es una zona con agua tan limpia, al menos no comparando con una zona donde el 90% de los registros de agua s sean verdes.<br><br>
Para evitar obtener este sesgo en la visualizacin de la calidad del agua por regiones, adems de obtener la moda, tambin computaremos el porcentaje que el valor de la moda representa del total de registros dentro de la misma regin; y este porcentaje lo utilizaremos para visualizar la intensidad del color del semforo por cada regin.

### **Conclusiones**

Realizamos un anlisis a los registros de aguas subterrnea de Mxico, con datos que nos proporciona la CONAGUA, para determinar si existe una relacin entre ubicacin geogrfica de los acuferos y su calidad de agua. Para este anlisis, tenemos la gran ventaja de que la CONAGUA ya maneja un sistema de semforo de tres colores (verde, amarillo y rojo) para determinar el nivel de calidad de agua de cada observacin de los acuferos.<br><br>
Ahora, los datos proporcionados por la CONAGUA nos muestra informacin de manera individual, mas no nos entrega resultados de cmo se encuentra la calidad de agua en diversas regiones del pas. Por lo tanto, para poder determinar la relacin del agua por regiones del pas, debemos agrupar los registros por regiones, y para ello nos hemos apoyado del algorimo K-means. Una vez que contamos con agrupaciones de registros de agua distribuidos a lo largo y ancho del pas, revisamos internamente el semforo de cada uno de los registros por agrupaciones, para asignar un valor de semforo a nivel macro, por agrupacin, lo cual nos da en trminos generales la respuesta sobre la relacin de una regin con la calidad del agua de la misma.<br><br>
Sin embargo, an con esta determinacin del semforo del agua por cada regin, seguimos sin conocer con qu nivel de certeza podemos mencionar si cierta regin del pas contiene agua limpia o agua contaminada. Es decir, este anlisis nos entrega resultados categricos, pero an no tenemos entendimiento sobre la escala de limpieza o contaminacin a nivel regional. Es por este motivo que recurrimos a computar el porcentaje que representa el valor de la moda del semforo de una regin, en relacin al total de registros dentro de la misma regin; de esta forma, contamos ya con un nivel de profundidad ms para comparar la calidad del agua de las diferentes regiones, incluso entre regiones cuyo valor del semforo es el mismo.<br><br>
Ya que logramos preparar los datos para obtener la relacin de la calidad del agua por regiones, qu observaciones logramos obtener del anlisis?
